```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('dEdXvsE.csv')
data['E(KeV)'] = data['E(MeV)']*1000
data['dEdX(MeV/mm)'] = data['dEdX(KeV/mm)']/1000
data.head()
```

```python
def exp_reg(x,y):
    model = np.polyfit(x, 1/y, 2)
    print(model)
    return lambda f: 1/((model[0]*f**2)+
                        (model[1]*f) + 
                        model[2])
```

```python
x = data['E(MeV)']
y = data['dEdX(KeV/mm)']
model = exp_reg(x, y)
E = np.linspace(0, 100, 200000)
dEdX = model(E)
plt.plot(x,y, '.')
plt.plot(E, dEdX, '-')
```

```python
delta_x = 0.1 #Step in milimeters
X = 0
Xs = []
energy = 100 #In MeV
energyChanges = []
while(energy>0):
    dEdX = model(energy)
    energyChange = dEdX*delta_x
    energyChanges.append(energyChange)
    Xs.append(X)
    X+=delta_x
    energy = energy-energyChange
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('dE(MeV)')
plt.ylabel('profundidad(mm)')
plt.plot(energyChanges, Xs)
plt.show()
```
